```python
import csv
import os

# Define a function to remove every other character from a string, excluding spaces
def process_string(s):
    result = []
    toggle = True  # To keep track of whether to keep the character or not
    for char in s:
        if char == ' ':
            result.append(char)
        elif toggle:
            result.append(char)
        toggle = not toggle
    return ''.join(result)

# Define input and output file paths
input_file = r'C:\Users\Smack\OneDrive\Documents\Websites\county to mineral.csv'
output_file = r'C:\Users\Smack\OneDrive\Documents\Websites\county to mineral output.csv'

# Check if the input file exists
if not os.path.isfile(input_file):
    print(f"Input file does not exist: {input_file}")
else:
    # Read the CSV file with a different encoding
    with open(input_file, mode='r', newline='', encoding='ISO-8859-1') as infile, open(output_file, mode='w', newline='', encoding='utf-8') as outfile:
        reader = csv.reader(infile, delimiter='\t')
        writer = csv.writer(outfile, delimiter='\t')
        
        # Process each row
        for row in reader:
            # Process only the first column
            if row and len(row) > 0:
                cell = row[0]
                row[0] = process_string(cell)
            writer.writerow(row)

```

```python
import pandas as pd

# Specify the input file path
input_file = r'C:\Users\Smack\OneDrive\Documents\Websites\website\mineral to county 2.csv'

# Load the CSV file
df = pd.read_csv(input_file)

# Ensure the DataFrame has at least two columns
if df.shape[1] < 2:
    df.insert(1, 'Wikipedia_Link', None)

# Function to create Wikipedia link
def create_wikipedia_link(mineral_name):
    if isinstance(mineral_name, str):
        mineral_name = mineral_name.strip()  # Remove leading/trailing spaces
        return f"https://m.wikipedia.org/wiki/{mineral_name.replace(' ', '_')}"
    else:
        return None  # Return None for non-string values

# Apply the function to the first column and save the result in the second column
df.iloc[:, 1] = df.iloc[:, 0].apply(create_wikipedia_link)

# Debugging: Print the DataFrame to check the results
print(df.head(10))  # Print first 10 rows to verify

# Specify the output file path
output_file = r'C:\Users\Smack\OneDrive\Documents\Websites\website\mineral to county 3.csv'

# Save the updated DataFrame to a new CSV file
df.to_csv(output_file, index=False)

print(f"Updated CSV file has been saved to {output_file}")
```
